# PS6 — Q1: Black–Scholes Greeks via Autograd

This notebook implements the Black–Scholes European call price in PyTorch and obtains delta, gamma, vega, theta, vanna, and volga using automatic differentiation.

The Black–Scholes call price is

$$C = S N(d_1) - K e^{-rT} N(d_2),$$

$$d_1 = \frac{\log(S/K)+(r+\frac{1}{2}\sigma^2)T}{\sigma\sqrt{T}}, \qquad d_2=d_1-\sigma\sqrt{T}.$$

Instead of manually deriving every Greek, we let PyTorch differentiate the pricing function.

```python
import sys
from pathlib import Path

tools_path = Path('../../.tools/python').resolve()
if tools_path.exists():
    sys.path.insert(0, str(tools_path))

from bs_autograd_greeks import autograd_call_greeks, run_ps6_q1

autograd_call_greeks(S0=100.0, K=100.0, T0=1.0, r0=0.02, sigma0=0.40)
```

Theta is reported using the common calendar-time convention $\Theta = -\partial C / \partial T$, where $T$ is time to maturity.
