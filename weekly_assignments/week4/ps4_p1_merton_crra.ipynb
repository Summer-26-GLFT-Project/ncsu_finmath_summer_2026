# PS4 — Problem 1: Merton CRRA

We solve the finite-horizon Merton portfolio problem with CRRA utility and verify the classic result: the optimal risky **portfolio share** is constant in wealth.

The wealth dynamics are

$$dW_t = (rW_t + \pi_t(\mu-r))dt + \pi_t\sigma dB_t,$$

where $\pi_t$ is the dollar amount in the risky asset. Under CRRA it is natural to use the risky share $\phi_t = \pi_t/W_t$.

For terminal utility

$$U(w)=\frac{w^{1-\gamma}}{1-\gamma},$$

the HJB first-order condition gives

$$\phi^* = \frac{\mu-r}{\gamma\sigma^2}.$$

So risk aversion controls the fraction of wealth invested in the risky asset, not a fixed dollar amount.

```python
from merton_crra import MertonCRRAParams, run_crra_experiment

params = MertonCRRAParams(mu=0.08, r=0.02, sigma=0.20, gamma=3.0, T=1.0)
share_check, summary = run_crra_experiment(params)
summary
```
